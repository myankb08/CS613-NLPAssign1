```python
parallel_dataset = {
    "hin_Deva": hindi_doc,
    "tel_Telu": telugu_doc
}
```

```python
def translate_docs(docs, src_lang, tgt_lang):
    translations = []
    for idx, doc in enumerate(docs):
        try:
            # Use preprocess_batch for batch of one
            preprocessed = ip.preprocess_batch([doc], src_lang=src_lang, tgt_lang=tgt_lang)
            if not preprocessed:
                print(f"Skipping empty doc at index {idx}")
                translations.append("")
                continue
        except Exception as e:
            print(f"Preprocessing error for index {idx}: {e}")
            translations.append("")
            continue
        
        inputs = tokenizer(preprocessed, return_tensors="pt", truncation=True, padding=True).to(DEVICE)
        forced_bos_token_id = tokenizer.convert_tokens_to_ids(f"<2{tgt_lang}>")
        with torch.no_grad():
            generated = model.generate(
                **inputs,
                num_beams=5,
                max_length=512,  # Use higher if doc is large
                forced_bos_token_id=forced_bos_token_id
            )
        decoded = tokenizer.batch_decode(generated, skip_special_tokens=True)
        try:
            # Use postprocess_batch for batch of one
            postprocessed = ip.postprocess_batch(decoded, lang=tgt_lang)
            translations.append(postprocessed[0] if postprocessed else decoded[0])
        except Exception as e:
            print(f"Postprocessing error for index {idx}: {e}")
            translations.append(decoded[0])
    return translations
```

```python
def evaluate_direction(src_lang, tgt_lang, dataset):
    """
    Evaluate translation performance from src_lang -> tgt_lang
    using chrF++ on a subset of the dataset.
    """
    src_sentences = dataset[src_lang]
    tgt_sentences = dataset[tgt_lang]

    # Translate
    preds = translate_docs(src_sentences, src_lang, tgt_lang)

    # Ensure preds is a flat list of strings
    preds = [p if isinstance(p, str) else "" for p in preds]

    if len(preds) == 0:
        print(f"⚠️ No predictions produced for {src_lang}->{tgt_lang}")
        return None

    from sacrebleu.metrics import CHRF
    chrf = CHRF(word_order=2)  # chrF++
    score = chrf.corpus_score(preds, [tgt_sentences]).score
    return score
```

```python
# Usage on your document lists:
preds = translate_docs(hindi_doc, src_lang="hin_Deva", tgt_lang="tel_Telu")

# Now call evaluate_direction by passing these predictions and references:
from sacrebleu.metrics import CHRF
chrf = CHRF(word_order=2)
score = chrf.corpus_score(preds, [telugu_doc]).score
print("chrF++ Hindi→Telugu:", score)

preds_tel = translate_docs(telugu_doc, src_lang="tel_Telu", tgt_lang="hin_Deva")

score_tel = chrf.corpus_score(preds_tel, [hindi_doc]).score
print("chrF++ Telugu→Hindi:", score_tel)
```
